Ćwiczenie 3

```python
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']
ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

```python
srednia = ceny.mean()
max_cena = ceny.max()
min_cena = ceny.min()

najdrozszy = nazwy[ceny.argmax()]
najtanszy = nazwy[ceny.argmin()]

print(srednia)
print(najdrozszy, max_cena)
print(najtanszy, min_cena)
```

```python
wartosci = ceny * ilosci_magazyn
laczna_wartosc = wartosci.sum()

print(wartosci)
print(laczna_wartosc)
```

```python
rabat = ceny * 0.85
vat = ceny * 1.23

ile_pow_1000 = (ceny > 1000).sum()

print(rabat)
print(vat)
print(ile_pow_1000)
```

```python
nazwy = np.array(nazwy)

maska_oceny = oceny_klientow >= 4.5
print(nazwy[maska_oceny])
print(ceny[maska_oceny])

maska_tanie = ceny < 200
print(nazwy[maska_tanie])
print(ilosci_magazyn[maska_tanie])
```

```python
maska_niskie = ilosci_magazyn < 30

print(nazwy[maska_niskie])

wartosci = ceny * ilosci_magazyn
wartosc_niskie = wartosci[maska_niskie].sum()

procent = (wartosc_niskie / wartosci.sum()) * 100

print(wartosc_niskie)
print(procent)
```

```python
okazje = (oceny_klientow >= 4.5) & (ceny < 1000)

print(nazwy[okazje])
print(ceny[okazje])

srednia_wysokie = ceny[oceny_klientow >= 4.0].mean()
srednia_niskie = ceny[oceny_klientow < 4.0].mean()

print(srednia_wysokie)
print(srednia_niskie)
```

Ćwiczenie 4

```python
import time
```

```python
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)
```

```python
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter() - start
start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter() - start
print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter() - start
start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter() - start
print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

```python
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter() - start
start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter() - start
print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

## Wnioski z benchmarku
- NumPy jest **19x** szybszy w mnożeniu
- NumPy jest **16x** szybszy w sumowaniu
- NumPy jest **23x** szybszy w filtrowaniu
- Wniosek: przy dużych danych **zawsze** używaj NumPy zamiast list
